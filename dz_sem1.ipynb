## Задача 2
#### На складе лежат разные фрукты в разном количестве. Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе/

```python
def count_fruits(*fruits):
    storage = dict(zip(fruits[::2], fruits[1::2]))
    return sum(storage.values())

print('Суммарное количество фруктов на складе:', count_fruits('яблоки', 24, 'бананы', 14, 'апельсины', 32))
```

## Задача 3
#### Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
#### **[100, 125, -90, 345, 655, -1, 0, 200]**
#### Используйте ***list comprehensions***

```python
lst = [100, 125, -90, 345, 655, -1, 0, 200]

lst = [val for val in lst if val >= 0]
total_spent = sum(lst)

print('Суммарные затраты на рекламу составили -', total_spent)
```

### Подзадача 1 
#### Найдите выручку компании в зависимости от месяца
#### Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
#### ***Используйте аннотирование типов.***

```python
from typing import List, Dict
import datetime

def calculate_revenue_by_month(dates: List[str], revenue: List[int]) -> Dict[str, int]:
    revenue_by_month = {}
    for i in range(len(dates)):
        date = datetime.datetime.strptime(dates[i], '%Y-%m-%d')
        month = date.strftime('%m-%Y')
        if month in revenue_by_month:
            revenue_by_month[month] += revenue[i]
        else:
            revenue_by_month[month] = revenue[i]
    return revenue_by_month

dates = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
revenue = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

print(calculate_revenue_by_month(dates, revenue))
```
